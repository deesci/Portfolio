### Columns content
Taking closer look at the contents of columns, we note from above that
- **One value** Some columns have mostly one value only. These hardly provide much info, like **offer_type** and **seller**. Both these are dominated by one value. These column hence can be removed.

### Odometer 


```python
autos.loc[:,'odometer']
```

as we can see above we need to **remove** `km` and "`,`" before converting text to number.

```python
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',','').astype(int)
```

```python
autos['odometer']
```

```python
autos.rename({'odometer': 'odometer_km'}, axis=1, inplace=True)
```

### Price
Similarly let us look at the contents of column `price` and understand how it needs to be **cleaned**.
We need to remove the `$` and `,`.

```python
autos['price']
```

```python
autos['price'] = autos['price'].str.replace('$','').str.replace(',','').astype(int)
```

```python
autos['price']
```

```python
autos.rename({'price': 'price_$'}, axis=1, inplace=True)
```

### Average price of topten

Having known the major brands, let us see their average prices of top ten brands. 

**Volkwagen is around , in dollar 6000, BMW  8500, opel  5000.**

    - audi   9212.9306621881,  
    - mercedes_benz   30025.542248978712,  
    - renault   2431.195698924731,  
    - peugeot   3065.611888111888,  
    - bmw   8493.857251184834,  
    - volkswagen   6600.382233404296,  
    - ford   7305.141843971631,  
    - opel   5281.321706161138,  
    - fiat   12566.541567695962

```python
brand_avgprice = {}
brand_avgmil = {}

for b in brand_topten:
    select_rows = new_autos[new_autos['brand'] == b]
    average_price = select_rows['price_$'].mean()
    brand_avgprice[b] = average_price
    average_mil = select_rows['odometer_km'].mean()
    brand_avgmil[b] = average_mil
    
print(brand_avgprice)
print(brand_avgmil)

    
```

```python
bmp_series = pd.Series(brand_avgprice)
bmm_series = pd.Series(brand_avgmil)
print(bmp_series)
print(bmm_series)
```

```python
top_ten_table = pd.DataFrame(bmp_series, columns=['mean_price'])
```
